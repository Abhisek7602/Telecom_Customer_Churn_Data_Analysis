#Data Cleaning

drop duplicates

```python
df.drop_duplicates(inplace=True)
```

Replace the null value with 0 and change the datatype of Totalcharge column

```python
df[df['TotalCharges']==' ']
```

```python
df['TotalCharges']=df['TotalCharges'].str.replace(' ', '0')
df['TotalCharges']=df['TotalCharges'].astype('float64')
```

#Data visualization

```python
plt.figure(figsize=(15,10),facecolor='yellow')
plt.suptitle('Customer Churn Analysis',fontsize=20)

plt.subplot(2,3,1)
plt.title('Count Of Customer By Churn')
a=sns.countplot(x='Churn',data=df,hue='Churn',palette='rainbow')
a.set_facecolor('violet')
for i in a.containers:
  a.bar_label(i)

plt.subplot(2,3,2)
plt.title('Percentage Of Churned Customer')
a=df.groupby('Churn').agg({'Churn':'count'})
plt.pie(a['Churn'],labels=a.index,autopct='%0.1f%%',explode=[0.1,0],colors=['red','blue'],shadow=True)

plt.subplot(2,3,3)
plt.title('Count Of Customer By Gender')
a=sns.countplot(x='gender',data=df,hue='Churn',palette='Set3')
a.set_facecolor('lightgreen')
for i in a.containers:
  a.bar_label(i)

plt.subplot(2,3,4)
plt.title('Count Of Customer By Gender')
a=sns.countplot(x='SeniorCitizen',data=df,hue='Churn',palette='dark:white')
a.set_facecolor('purple')
for i in a.containers:
  a.bar_label(i)

plt.subplot(2,3,5)
a=sns.histplot(y='tenure',data=df,bins=72,hue='Churn')
a.set_facecolor('pink')
plt.title('Count Of Customer By Tenure')

plt.subplot(2,3,6)
plt.title('Count Of Customer By Contract')
a=sns.countplot(x='Contract',data=df,hue='Churn',palette='dark:white')
a.set_facecolor('crimson')
for i in a.containers:
  a.bar_label(i)

plt.tight_layout()
plt.show()
```

1.Overall Churn Rate:-
According to the data 5174 customers have stayed while 1869 customers have curned.
Around 26.5% of customers have churned, while 73.5% have stayed.

2.Churn by Gender:-
Female: 939 churned (26.9%) vs 2549 retained.
Male: 930 churned (26.1%) vs 2625 retained.
Churn rate is almost the same across genders.

3.Churn by Senior Citizen Status:-
Non-Senior Citizens: 1393 churned (23.6%) vs 4508 retained.
Senior Citizens: 476 churned (41.6%) vs 666 retained.
Non-Senior Citizens have a much higher churn rate than Seniors.

4.Churn by Tenure:-
Customers with shorter tenure (0–12 months) show the highest churn.
Longer-tenure customers (2+ years) tend to stay.

5.Churn by Contract Type:-
Month-to-Month: 1655 churned vs 2220 retained → highest churn.
One Year Contract: 166 churned vs 1307 retained → low churn.
Two Year Contract: 48 churned vs 1647 retained → lowest churn.
Customers with longer contracts are far more loyal.

6.Phone Service:-

7.Multiple Lines:-Customers without multiple lines churn more than those with multiple lines.Having additional lines seems to reduce churn slightly.

8.Internet Service:-Fiber optic users show the highest churn compared to DSL.
Customers without internet service show very low churn.

9.Online Security:-Lack of online security strongly correlates with higher churn.Customers using online security services churn much less.

10.Online Backup:-Customers without backup services have higher churn compared to those who use backups.Online backup availability helps retain customers.

11.Device Protection:-Customers with device protection churn less.Not having device protection increases churn.

12.Tech Support:-Lack of tech support is a major driver of churn.Customers with tech support churn less significantly.

13.Streaming TV:-Customers with streaming TV show moderate churn.Lack of streaming TV slightly reduces churn but not as much as service-related features.

14.Streaming Movies:-Customers with streaming movies show slightly higher churn than those without.
